```python
# Importações gerais
import pandas as pd
```

```python
# Importando a base estática de jogadores da NBA
from nba_api.stats.static import players
# Pega uma lista de dicionários, cada um um jogador
nba_players = players.get_players()
```

```python
###### TROCAR O NOME DO JOGADOR AQUI ##########
nome = 'LeBron James'
# Acha o dicionário correspondente ao jogador desejado
jogador = [player for player in nba_players
                   if player['full_name'] == nome][0]
jogador
```

```python
# Importando a API que acha os jogos
from nba_api.stats.endpoints import leaguegamefinder
# Procurando os jogos que o jogador participou
gamefinder = leaguegamefinder.LeagueGameFinder(player_id_nullable=jogador['id'])
jogos = gamefinder.get_data_frames()[0]
jogos.head()
```

```python
#A coluna de temporada tem um identificador numérico de tipo logo antes. Os ids de temporada regular começam com o número 2
jogos.groupby(jogos.SEASON_ID)[['GAME_ID']].count()
```

```python
# Vamos fazer esse corte e manter apenas o ano
jogos = jogos[jogos['SEASON_ID'].str[0] == '2']
jogos['SEASON_ID'] = jogos['SEASON_ID'].str[-4:]
```

```python
# Trocando o nome da coluna de temporada
colunas = list(jogos.columns)
colunas[0] = 'SEASON'
jogos.columns = colunas
```

```python
# Criando uma coluna para dizer se o time jogou em casa
jogos['HOME'] = jogos['MATCHUP'].str[4:5] == 'v'
```

```python
# Dimensões dos dados
jogos.shape
```

```python
# Tipos das colunas
jogos.info()
```

```python
# Transformando IDs em numéricos
jogos['SEASON'] = pd.to_numeric(jogos['SEASON'])
jogos['GAME_ID'] = pd.to_numeric(jogos['GAME_ID'])
```

```python
# Conferindo
jogos.tail().T
```

```python
arquivo = nome+'.csv'
jogos.to_csv(arquivo)
```
